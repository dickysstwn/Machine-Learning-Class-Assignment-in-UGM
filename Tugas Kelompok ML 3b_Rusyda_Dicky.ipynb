## **d.**
Lakukan pra-pemrosesan data (*data preprocessing*) dan normalisasi fitur (*feature scaling*) yang diperlukan untuk kumpulan data.

```python
# Mendownload data train dan test
train_url = 'https://drive.google.com/file/d/14DQUwBlSWD43rCAp_s_pWRyv77yAIMfs/view?usp=sharing'
test_url = 'https://drive.google.com/file/d/1D4_nw7UABmTJI4EtRXUsfT4pis5AqSA2/view?usp=sharing'

train_path = 'train.csv'
test_path = 'test.csv'

# Mengunduh data train
train_file_id = train_url.split('/')[-2]
!gdown --id $train_file_id -O $train_path

# Mengunduh data test
test_file_id = test_url.split('/')[-2]
!gdown --id $test_file_id -O $test_path

# Membaca data train dan test menggunakan pandas
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# Memisahkan kolom target (SalePrice) dari data train
target = train_data['SalePrice']
train_data.drop('SalePrice', axis=1, inplace=True)

# Menggabungkan data train dan test untuk pra-pemrosesan yang konsisten
combined_data = pd.concat([train_data, test_data], axis=0)

# Menghapus kolom yang memiliki banyak nilai yang hilang (missing values)
threshold = len(combined_data) * 0.8  # Hapus kolom jika lebih dari 80% nilai hilang
combined_data.dropna(thresh=threshold, axis=1, inplace=True)

# Mengisi nilai hilang pada kolom numerik dengan nilai rata-rata
numeric_cols = combined_data.select_dtypes(include=['float64', 'int64']).columns
combined_data[numeric_cols] = combined_data[numeric_cols].fillna(combined_data[numeric_cols].mean())

# Mengisi nilai hilang pada kolom kategorikal dengan nilai yang paling sering muncul
categorical_cols = combined_data.select_dtypes(include=['object']).columns
combined_data[categorical_cols] = combined_data[categorical_cols].fillna(combined_data[categorical_cols].mode().iloc[0])

# Mengubah kolom kategorikal menjadi numerik dengan one-hot encoding
combined_data = pd.get_dummies(combined_data)

# Memisahkan kembali data train dan test setelah pra-pemrosesan
train_data = combined_data[:len(train_data)]
test_data = combined_data[len(train_data):]

# Melakukan feature scaling pada data train dan test menggunakan StandardScaler
scaler = StandardScaler()
train_data_scaled = scaler.fit_transform(train_data)
test_data_scaled = scaler.transform(test_data)
```

## **e.**
Buatlah berbagai **model regresi** secara individual dan sesuaikan *hyperparameter* ***(hyperparameter tuning)*** untuk setiap model. Setelah mendapatkan *hyperparameter* terbaik, evaluasi model dengan dataset *testing* menggunakan metrik *Root Mean Square Error* (RMSE) untuk membandingkan hasil prediksi.
* *Linear Regression*
* *Lasso Regression*
* *ElasticNet Regression*
* *Ridge Regression*


```python
# Memisahkan dataset training menjadi training set dan validation set
X_train, X_val, y_train, y_val = train_test_split(train_data_scaled, target, test_size=0.2, random_state=42)
```

### ***Linear Regression***

```python
# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
linear_reg_pred = linear_reg.predict(X_val)
linear_reg_rmse = np.sqrt(mean_squared_error(y_val, linear_reg_pred))
```

```python
# Menampilkan hasil evaluasi model
print("Linear Regression RMSE:", linear_reg_rmse)
```

### ***Lasso Regression***

```python
# Lasso Regression
lasso_reg = Lasso()
lasso_params = {'alpha': [0.001, 0.01, 0.1, 1, 10]}
lasso_grid = GridSearchCV(lasso_reg, lasso_params, scoring='neg_root_mean_squared_error', cv=5)
lasso_grid.fit(X_train, y_train)
lasso_reg_best = lasso_grid.best_estimator_
lasso_reg_pred = lasso_reg_best.predict(X_val)
lasso_reg_rmse = np.sqrt(mean_squared_error(y_val, lasso_reg_pred))
```

```python
# Menampilkan hasil evaluasi model
print("Lasso Regression RMSE:", lasso_reg_rmse)
```

### ***ElasticNet Regression***

```python
# ElasticNet Regression
elastic_net = ElasticNet()
elastic_params = {'alpha': [0.001, 0.01, 0.1, 1, 10], 'l1_ratio': [0.2, 0.4, 0.6, 0.8]}
elastic_grid = GridSearchCV(elastic_net, elastic_params, scoring='neg_root_mean_squared_error', cv=5)
elastic_grid.fit(X_train, y_train)
elastic_net_best = elastic_grid.best_estimator_
elastic_net_pred = elastic_net_best.predict(X_val)
elastic_net_rmse = np.sqrt(mean_squared_error(y_val, elastic_net_pred))
```

```python
# Menampilkan hasil evaluasi model
print("ElasticNet Regression RMSE:", elastic_net_rmse)
```

### ***Ridge Regression***

```python
# Ridge Regression
ridge_reg = Ridge()
ridge_params = {'alpha': [0.001, 0.01, 0.1, 1, 10]}
ridge_grid = GridSearchCV(ridge_reg, ridge_params, scoring='neg_root_mean_squared_error', cv=5)
ridge_grid.fit(X_train, y_train)
ridge_reg_best = ridge_grid.best_estimator_
ridge_reg_pred = ridge_reg_best.predict(X_val)
ridge_reg_rmse = np.sqrt(mean_squared_error(y_val, ridge_reg_pred))
```

```python
# Menampilkan hasil evaluasi model
print("Ridge Regression RMSE:", ridge_reg_rmse)
```

### **Evaluasi model dengan metrik Root Mean Square Error (RMSE)**

```python
# Menampilkan hasil evaluasi model
print("Linear Regression RMSE:", linear_reg_rmse)
print("Lasso Regression RMSE:", lasso_reg_rmse)
print("ElasticNet Regression RMSE:", elastic_net_rmse)
print("Ridge Regression RMSE:", ridge_reg_rmse)
```

## **f.**
Cobalah berbagai pemodelan *ensemble learning* menggunakan beberapa model dengan hyperparameter yang sudah di-tuning. Evaluasi pemodelan *ensemble* dengan dataset *testing* menggunakan metrik *RMSE*.
* *Bagging Regressor*

### ***Bagging Regressor***

```python
# Bagging Regressor
bagging_reg = BaggingRegressor(n_estimators=100, random_state=42)
bagging_reg.fit(X_train, y_train)
bagging_pred = bagging_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Bagging Regressor
bagging_rmse = np.sqrt(-cross_val_score(bagging_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Bagging Regressor RMSE:", np.mean(bagging_rmse))
```

### ***Forests of Randomized Trees (Random Forest Regressor, Extra Tress Regressor)***

```python
# Random Forest Regressor
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train, y_train)
rf_pred = rf_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Random Forest Regressor
rf_rmse = np.sqrt(-cross_val_score(rf_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Random Forest Regressor RMSE:", np.mean(rf_rmse))
```

```python
# Extra Trees Regressor
extra_trees_reg = ExtraTreesRegressor(n_estimators=100, random_state=42)
extra_trees_reg.fit(X_train, y_train)
extra_trees_pred = extra_trees_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Extra Trees Regressor
extra_trees_rmse = np.sqrt(-cross_val_score(extra_trees_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Extra Trees Regressor RMSE:", np.mean(extra_trees_rmse))
```

### ***Boosting Regressor (AdaBoost Regressor, Gradient Boosting Regressor, XGB Regressor)***

```python
# AdaBoost Regressor
adaboost_reg = AdaBoostRegressor(n_estimators=100, random_state=42)
adaboost_reg.fit(X_train, y_train)
adaboost_pred = adaboost_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model AdaBoost Regressor
adaboost_rmse = np.sqrt(-cross_val_score(adaboost_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("AdaBoost Regressor RMSE:", np.mean(adaboost_rmse))
```

```python
# Gradient Boosting Regressor
gradboost_reg = GradientBoostingRegressor(n_estimators=100, random_state=42)
gradboost_reg.fit(X_train, y_train)
gradboost_pred = gradboost_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Gradient Boosting Regressor
gradboost_rmse = np.sqrt(-cross_val_score(gradboost_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Gradient Boosting Regressor RMSE:", np.mean(gradboost_rmse))
```

```python
# XGB Regressor
xgb_reg = xgb.XGBRegressor(n_estimators=100, random_state=42)
xgb_reg.fit(X_train, y_train)
xgb_pred = xgb_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model XGB Regressor
xgb_rmse = np.sqrt(-cross_val_score(xgb_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("XGB Regressor RMSE:", np.mean(xgb_rmse))
```

### ***Voting Regressor***

```python
# Voting Regressor
voting_reg = VotingRegressor([('rf', rf_reg), ('gradboost', gradboost_reg), ('xgb', xgb_reg)])
voting_reg.fit(X_train, y_train)
voting_pred = voting_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Voting Regressor
voting_rmse = np.sqrt(-cross_val_score(voting_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Voting Regressor RMSE:", np.mean(voting_rmse))
```

### ***Stacking Regressor***

```python
# Stacking Regressor
stacking_reg = StackingRegressor([('rf', rf_reg), ('gradboost', gradboost_reg), ('xgb', xgb_reg)], final_estimator=LinearRegression())
stacking_reg.fit(X_train, y_train)
stacking_pred = stacking_reg.predict(test_data_scaled)
```

```python
# Menampilkan RMSE untuk model Stacking Regressor
stacking_rmse = np.sqrt(-cross_val_score(stacking_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Stacking Regressor RMSE:", np.mean(stacking_rmse))
```

### **Evaluasi model dengan metrik Root Mean Square Error (RMSE)**

```python
# Menampilkan RMSE untuk masing-masing model ensemble
bagging_rmse = np.sqrt(-cross_val_score(bagging_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Bagging Regressor RMSE:", np.mean(bagging_rmse))

rf_rmse = np.sqrt(-cross_val_score(rf_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Random Forest Regressor RMSE:", np.mean(rf_rmse))

extra_trees_rmse = np.sqrt(-cross_val_score(extra_trees_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Extra Trees Regressor RMSE:", np.mean(extra_trees_rmse))

adaboost_rmse = np.sqrt(-cross_val_score(adaboost_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("AdaBoost Regressor RMSE:", np.mean(adaboost_rmse))

gradboost_rmse = np.sqrt(-cross_val_score(gradboost_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Gradient Boosting Regressor RMSE:", np.mean(gradboost_rmse))

xgb_rmse = np.sqrt(-cross_val_score(xgb_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("XGB Regressor RMSE:", np.mean(xgb_rmse))

voting_rmse = np.sqrt(-cross_val_score(voting_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Voting Regressor RMSE:", np.mean(voting_rmse))

stacking_rmse = np.sqrt(-cross_val_score(stacking_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=5))
print("Stacking Regressor RMSE:", np.mean(stacking_rmse))
```

## **g.**
Analisis perbedaan hasil antara berbagai model regresi secara individual dan model *ensemble learning*

Dalam analisis hasil antara model regresi secara individual dan model ensemble learning, kita dapat melihat beberapa perbedaan yang terjadi diantaranya:

1. Model ensemble learning cenderung memberikan performa yang lebih baik daripada model regresi individual. Karena ensemble learning menggabungkan prediksi dari beberapa model, model ini dapat mengurangi bias dan varian yang terkait dengan model individual. Hasilnya, model ensemble learning memiliki kecenderungan untuk memberikan prediksi yang lebih akurat dan stabil.
2. Model ensemble learning memiliki kecenderungan yang lebih baik dalam menangani outliers dan data yang tidak teratur. Dengan menggunakan gabungan prediksi dari beberapa model, ensemble learning dapat menyerap ketidaksempurnaan dari setiap model individual dan memberikan hasil yang lebih robust.
3. Ensemble learning memiliki kecenderungan yang lebih baik dalam generalisasi ke data baru. Karena ensemble learning mempelajari relasi antara model individual dan memadukan prediksi mereka, model ini dapat mengatasi overfitting dan mampu menghasilkan prediksi yang lebih umum dan berlaku untuk berbagai situasi.
4. Dalam ensemble learning, jika satu atau beberapa model individual memberikan prediksi yang buruk, prediksi keseluruhan masih bisa menjadi baik. Ini menjadikan ensemble learning lebih andal dalam menghadapi kegagalan model individual atau data yang buruk.
5. Ensemble learning cenderung memiliki kompleksitas yang lebih tinggi daripada model regresi individual. Karena melibatkan beberapa model, ensemble learning memerlukan waktu dan sumber daya komputasi yang lebih banyak untuk melatih dan menghasilkan prediksi.
6. Model regresi individual cenderung lebih mudah diinterpretasikan karena kita dapat melihat koefisien dan pengaruh masing-masing fitur terhadap prediksi. Di sisi lain, ensemble learning, terutama dengan menggunakan metode seperti stacking, dapat menjadi lebih sulit untuk diinterpretasikan karena melibatkan kombinasi dari beberapa model.

**Secara keseluruhan, model ensemble learning dapat memberikan hasil yang lebih baik dibandingkan dengan model regresi individual**. Namun, kompleksitas yang ada dapat menjadi tantangan dalam penggunaan ensemble learning. Penting untuk mempertimbangkan kebutuhan dan tujuan spesifik saat memilih antara kedua pendekatan ini.
